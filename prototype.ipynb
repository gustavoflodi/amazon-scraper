```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time

# Initialize Selenium WebDriver
options = webdriver.ChromeOptions()
options.add_argument('--headless')  # Run in headless mode for faster execution
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

reviews = {}
stars = ['one_star', 'two_star', 'three_star', 'four_star', 'five_star']
product = 'B095749V5V'

for star in stars:
    comments = list()
    for num_page in range(1, 10):
        
        url = f'https://www.amazon.com.br/product-reviews/{product}/ref=cm_cr_arp_d_viewopt_sr?filterByStar={star}&pageNumber={num_page}'
        print('Scraping this url:', url)
        
        driver.get(url)
        time.sleep(5)  # Adjust based on how fast the page loads
        
        # Locate all review elements
        reviews_elements = driver.find_elements(By.XPATH, '//span[@data-hook="review-body"]')
        
        if len(reviews_elements) == 0:
            print(f"No reviews found for {star} page {num_page}")
            break
        
        for review in reviews_elements:
            comments.append(review.text.strip())
    
    reviews[star] = comments

driver.quit()

# Print the collected reviews
for star, comments in reviews.items():
    print(f"\n{star} reviews:")
    for comment in comments:
        print(comment)
```

```python
import re
from collections import Counter
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import nltk
nltk.download('stopwords')
from textblob import TextBlob
from langdetect import detect
from googletrans import Translator
```

```python
from nltk.util import ngrams
```

```python
# Initialize the translator
translator = Translator()

# Set of stop words (in English for now)
stop_words = set(stopwords.words('english'))

def preprocess_text(text, lang='en'):
    # Lowercase, remove punctuation and stop words
    text = text.lower()
    text = re.sub(r'[^\w\s]', '', text)
    words = text.split()
    if lang == 'en':
        words = [word for word in words if word not in stop_words]
    return words

def detect_and_translate(text):
    try:
        detected_lang = detect(text)
        if detected_lang != 'en':
            translated = translator.translate(text, src=detected_lang, dest='en')
            return translated.text, 'en'
        else:
            return text, detected_lang
    except:
        return text, 'unknown'

def get_most_common_words(reviews_dict):
    star_common_words = {}
    for star, reviews in reviews_dict.items():
        word_counter = Counter()
        for review in reviews:
            translated_review, lang = detect_and_translate(review)
            words = preprocess_text(translated_review, lang)
            word_counter.update(words)
        star_common_words[star] = word_counter.most_common(10)
    return star_common_words

def perform_sentiment_analysis(reviews_dict):
    sentiment_scores = {}
    for star, reviews in reviews_dict.items():
        sentiment = 0
        count = 0
        for review in reviews:
            translated_review, _ = detect_and_translate(review)
            blob = TextBlob(translated_review)
            sentiment += blob.sentiment.polarity
            count += 1
        sentiment_scores[star] = sentiment / count if count > 0 else 0
    return sentiment_scores

# Most common words
star_common_words = get_most_common_words(reviews)
print("Most Common Words for Each Star:")
for star, common_words in star_common_words.items():
    print(f"\n{star.capitalize()} Reviews:")
    for word, count in common_words:
        print(f"{word}: {count}")

# Sentiment analysis
sentiment_scores = perform_sentiment_analysis(reviews)
print("\nSentiment Scores:")
for star, score in sentiment_scores.items():
    print(f"{star.capitalize()}: {score}")
```
